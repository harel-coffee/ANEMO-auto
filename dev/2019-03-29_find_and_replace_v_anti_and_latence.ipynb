```python
import pickle
import os
```

```python
liste = []
for path, subdirs, files in os.walk('../../'):
    for name in files:
        if name[-4:] == '.pkl' :
            liste.append(os.path.join(path, name))
            #print(os.path.join(path, name))
```

```python
list_NON = []
for file in liste :
    print('\n',file)
    
    if file[-4:] == '.pkl' :
        with open(file, 'rb') as fichier :
            try :
                a = 0
                x = pickle.load(fichier, encoding='latin1')
                print(x.keys())
            
                if 'v_anti' in x.keys():
                    x['a_anti'] = x['v_anti']
                    if (x['a_anti'] == x['v_anti']) : del(x['v_anti']) ; a=1
                if 'latence' in x.keys():
                    x['latency'] = x['latence']
                    if (x['latency'] == x['latence']) : del(x['latence']) ; a=1
                if 'old_latence' in x.keys():
                    x['old_latency'] = x['old_latence']
                    if (x['old_latency'] == x['old_latence']) : del(x['old_latence']) ; a=1
                print(x.keys())
                print(a)
                if a ==1 :
                    # saves the fit parameters
                    with open(file, 'wb') as fichier:
                        f = pickle.Pickler(fichier)
                        f.dump(x)
            except : list_NON.append(file) ;  print('NON')
```
